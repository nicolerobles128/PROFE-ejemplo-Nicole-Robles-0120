```python
class automovil:
    def __init__(self, marca, placa, precio):
        self.marca = marca
        self.placa = placa
        self.precio = precio

    def leerDatos(self):
        self.marca = input("Ingrese la marca: ")
        self.placa = input("Ingrese la placa: ")
        self.precio = int(input("Ingrese el precio: "))


    def mostrarDatos(self):
        print("Marca: ", self.marca)
        print("Placa: ", self.placa)
        print("Precio: ", self.precio)

# Creando los objetos carro1, carro2, carro3
carro1 = automovil("","","")
carro2 = automovil("","","")
carro3 = automovil("","","")

print("Carro Numero 1")# creando el objeto carro1
carro1.leerDatos()
carro1.mostrarDatos()
print("-*-*-*-*-*-*-*-*-*-*-*")

print("Carro Numero 2")# creando el objeto carro2
carro2.leerDatos()
carro2.mostrarDatos()
print("-*-*-*-*-*-*-*-*-*-*-*")

print("Carro Numero 3")# creando el objeto carro3
carro3.leerDatos()
carro3.mostrarDatos()
print("Software dasarrollado por Nicole Robles NL = 46")
```
